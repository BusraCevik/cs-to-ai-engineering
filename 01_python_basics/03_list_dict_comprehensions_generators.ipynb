**Short Description**
Comprehensions and generators provide concise, readable, and efficient ways to create lists, dictionaries, and iterators in Python.

## Traditional For Loop

```python
squares = []
for x in range(5):
    squares.append(x**2)

print(squares)  # [0, 1, 4, 9, 16]
```

## List Comprehension

```python
squares = [x**2 for x in range(5)]
print(squares)  # [0, 1, 4, 9, 16]
```

## The Syntax

```python
newlist = [expression for item in iterable if condition == True]
```

## List Comprehension With Condition


```python
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print(even_squares)  # [0, 4, 16, 36, 64]
```

## Dict Comprehension

```python
squared_dict = {x: x**2 for x in range(5)}
print(squared_dict)  # {0:0, 1:1, 2:4, 3:9, 4:16}
```

## Dict Comprehension With condition:

```python
squared_even_dict = {x: x**2 for x in range(10) if x % 2 == 0}
print(squared_even_dict)  # {0:0, 2:4, 4:16, 6:36, 8:64}
```

## Generators

* lazy evaluation â†’ memory efficient
* Useful for large datasets or streaming data

```python
gen = (x**2 for x in range(5))
print(next(gen))  # 0
print(next(gen))  # 1

for val in gen:
    print(val)
# Output: 4 9 16
```
