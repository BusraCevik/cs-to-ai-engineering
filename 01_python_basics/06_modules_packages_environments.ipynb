## Python Implementation / Creating a Module

```python
# helper.py
def greet(name):
    return f"Hello, {name}!"

# main.py
from helper import greet

print(greet("Busra"))
```

## Python Implementation / Creating a Package

```python
# utils/__init__.py  (can be empty)
# utils/math_utils.py
def add(a, b):
    return a + b

# main.py
from utils.math_utils import add
print(add(5, 3))
```

## Python Implementation / Using Virtual Environment

```python
# create environment
python -m venv venv

# activate environment (Windows)
venv\Scripts\activate

# activate environment (Linux/Mac)
source venv/bin/activate

# install packages
pip install requests
pip freeze > requirements.txt
pip install -r requirements.txt  # install all dependencies
```
