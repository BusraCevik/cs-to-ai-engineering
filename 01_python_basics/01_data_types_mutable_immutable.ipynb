## 1️⃣ Description
  Python data types are divided into immutable and mutable; mutable can be changed, immutable cannot.

---

## 2️⃣ Python Concepts / Notes
- **Variables:** storing data
- **Data Types:** int, float, str, bool
- **Basic Operators:** +, -, *, /, %, //, **
- **Input/Output:** input(), print()
- **Comments:** use `#` for single line
- **Functions:** defining and calling

---

* Immutable: int, float, str, tuple
* Mutable: list, dict, set



```python
x = 10  # int, immutable
y = [1,2,3]  # list, mutable
```

```python
# Immutable
x = 5
# x[0] = 3  -> Error

# Mutable
y = [1, 2, 3]
y[0] = 9  # works
print(y)  # [9,2,3]
```

```python
#Tips: Lists inside tuples can be changed
a = (1, [2,3])
a[1][0] = 99
print(a)  # (1,[99,3])
```

## Note ->
   Time / Space Complexity: O(1)
